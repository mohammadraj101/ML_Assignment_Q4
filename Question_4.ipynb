```python
import pandas as pd
import numpy as np


file_path = 'WineQT.csv'  
data = pd.read_csv(file_path)
data.describe().transpose()[["min","max"]]
```

```python

data['quality'] = np.where(data['quality'] > 6, 1, 0)

columns_to_normalize = ['fixed acidity', 'volatile acidity', 'citric acid', 
                         'residual sugar', 'chlorides', 'free sulfur dioxide', 
                         'total sulfur dioxide', 'density', 'pH', 'sulphates', 
                         'alcohol']

# Manually normalize each column
for column in columns_to_normalize:
    # Calculating min and max
    min_val = data[column].min()
    max_val = data[column].max()
    
    # Applying min-max normalization 
    data[column] = (data[column] - min_val) / (max_val - min_val)

#to save the changes in csv file
transformed_file_path = 'WineQT_Modified.csv'  
data.to_csv(transformed_file_path, index=False)
```
